```python
import matplotlib.pyplot as plt
%matplotlib inline
from ethno import EthnoModel, EthnoAgent
import numpy as np
```

```python
m = EthnoModel(5,50,50,1,0.005, 1, [0b00, 0b01, 0b10, 0b11], max_iters=2000)
```

```python
m.run_model()
```

```python
data = m.datacollector.get_model_vars_dataframe()
```

```python
ax = data[["Ethnocentric","Humanitarian", "Selfish", "Traitor"]].plot()
ax.set_title("Population and Behavior Over Time")
ax.set_xlabel("Step")
ax.set_ylabel("Number of Agents")
_ = ax.legend(bbox_to_anchor=(1.35, 1.025))
```

```python
misper_data = data[["Misperception Stats"]]
# there's probably a better way to do this but i don't know much about pandas?
inds = []
means = []
medians = []
std_devs = []

for i, row in misper_data.iterrows():
    info = row.tolist()[0]
    inds.append(i)
    means.append(info[0])
    medians.append(info[1])
    std_devs.append(info[2])
    
inds = np.array(inds)
means = np.array(means)
medians = np.array(medians)
std_devs = np.array(std_devs)
```

```python
fig, ax = plt.subplots(1)
ax.plot(inds, means, lw=2, label='mean misperception', color='blue')
ax.plot(inds, medians, lw=2, label='median misperception', color='red')
ax.fill_between(inds, means+std_devs, means-std_devs, facecolor='blue', alpha=0.5)
ax.set_title('Misperception over time')
ax.legend(loc='upper left')
ax.set_xlabel('num steps')
ax.set_ylabel('misperception')
ax.grid()
```
