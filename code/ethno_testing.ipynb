```python
import matplotlib.pyplot as plt
%matplotlib inline
from ethno import EthnoModel, EthnoAgent
import numpy as np
```

```python
m_default = EthnoModel(100, 50, 50, 1, 0.005, 1, [0b00, 0b01, 0b10, 0b11])
```

```python
m_start_0 = EthnoModel(100, 50, 50, 1, 0.005, 1, [0b00, 0b01, 0b10, 0b11])
for agent in m_start_0.schedule.agents:
    agent.misperception = 0
```

```python
m_start_1 = EthnoModel(100, 50, 50, 1, 0.005, 1, [0b00, 0b01, 0b10, 0b11])
for agent in m_start_1.schedule.agents:
    agent.misperception = 1
```

```python
%time m_default.run_model()
```

```python
%time m_start_0.run_model()
```

```python
%time m_start_1.run_model()
```
