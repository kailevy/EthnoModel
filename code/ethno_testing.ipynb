```python
import matplotlib.pyplot as plt
%matplotlib inline
from ethno import EthnoModel, EthnoAgent
```

```python
m = EthnoModel(5,50,50,1,0.005,1,[0b00, 0b01, 0b10, 0b11], max_iters=100)
```

```python
m.run_model()
```

```python
data = m.datacollector.get_model_vars_dataframe()
```

```python
ax = data[["Ethnocentric","Humanitarian", "Selfish", "Traitor"]].plot()
ax.set_title("Population and Behavior Over Time")
ax.set_xlabel("Step")
ax.set_ylabel("Number of Agents")
_ = ax.legend(bbox_to_anchor=(1.35, 1.025))
```

```python
BEHAVIOR_KEY = {0b11:"H", 0b10:"E", 0b01:"T", 0b00:"S"}
COLOR_KEY = {1:"#ff7c7a", 2:"#81ff7a", 3:"#7aadff", 4:"#faff7a"}
data_keys = [str(i)+BEHAVIOR_KEY[j] for i in range(1,5) for j in range(4)]
print(data_keys)
print(data[data_keys])
```

```python
styles = {'S': '-', 'T': '--', 'H': '-.', 'E':':'}
colors = {'1': 'r', '2': 'b', '3': 'g', '4': 'c'}
plot_style = [colors[key[0]] + styles[key[1]] for key in data_keys]
ax = data[data_keys].plot(style=plot_style)
ax.set_title("Population and Behavior Over Time")
ax.set_xlabel("Step")
ax.set_ylabel("Number of Agents")
_ = ax.legend(bbox_to_anchor=(1.35, 1.025))
```
